1.What are data structures, and why are they important?
  >>What Are Data Structures?
A data structure is a way of organizing, storing, and managing data to perform operations efficiently. It defines the relationship between data and the operations that can be performed on it.

  There are two main types of data structures:

- Linear Data Structures – Elements are arranged sequentially.
Examples: Arrays, Linked Lists, Stacks, Queues
- Non-Linear Data Structures – Elements are arranged in a hierarchical or interconnected manner.
Examples: Trees, Graphs, Hash Tables, Heaps

  Why Are Data Structures Important?
- Efficiency – They help in efficiently managing and retrieving data, reducing time and space complexity.
- Scalability – Well-chosen data structures allow applications to handle large amounts of data effectively.
- Better Problem Solving – Many algorithms rely on specific data structures for optimization (e.g., Dijkstra’s algorithm uses priority queues).
- Code Maintainability – Organizing data properly makes code more readable and maintainable.
- Optimized Resource Usage – Proper data structure selection leads to better memory and processing power utilization.

4.Describe how dictionaries store data ?
  >>Python dictionaries use a hash table with the following process:

  Hashing the Key

- The dictionary computes a hash value (a unique number) for the key using Python’s built-in hash() function.
Finding an Index

- The hash value determines the index in the internal array where the key-value pair is stored.
If two different keys generate the same index (collision), Python uses a collision resolution technique (e.g., open addressing).
Retrieval Process

- When accessing a value using dict[key], Python:
Computes the hash of the key.
Locates the corresponding index.
Returns the associated value in constant time O(1) (on average).

5.Why might you use a set instead of a list in Python?
  >>When to Use a Set Instead of a List
✅ 1. Removing Duplicates Efficiently
Since sets automatically remove duplicates, they are useful for cleaning data.
✅ 2. Faster Membership Checking (in operator)
Sets: O(1) average time complexity (uses a hash table).
Lists: O(n) worst-case time complexity (requires full traversal).
✅ 3. Mathematical Set Operations
Sets support union, intersection, and difference operations.
3. When NOT to Use a Set
❌ Order Matters → Use a list (list maintains insertion order).
❌ Duplicate Elements Are Needed → Lists allow duplicates.
❌ Index-Based Access Required → Sets don’t support indexing (set[0] is invalid).

8. What is a hash table, and how does it relate to dictionaries in Python?
  >> A hash table (also known as a hash map) is a data structure that stores key-value pairs using a hash function to compute an index (or address) where values are stored. It allows for fast lookups, insertions, and deletions in average O(1) time complexity.Python dictionaries (dict) are implemented using hash tables. Each dictionary key is hashed, and the hash value determines where the value is stored in memory.

10. Explain why strings are immutable in Python?
  >>Reasons for String Immutability
- Memory Efficiency (String Interning)
Python optimizes memory usage by reusing immutable strings.
If two variables have the same string value, Python points them to the same memory location.
- Hashing and Dictionary Keys
Strings are hashable, allowing them to be used as keys in dictionaries.
Hashing requires fixed (unchangeable) values to ensure consistent lookup times.
- Thread Safety
Since strings cannot be modified, multiple threads can safely share them without synchronization issues.
- Security Benefits
String immutability prevents accidental or malicious modification of critical data, such as passwords.

14. How does the “in” keyword work differently for lists and dictionaries?
  >>The in keyword is used to check for membership in both lists and dictionaries, but its behavior differs between these two data structures.
  - Checking Membership in a List
When you use the in keyword with a list, Python checks if the value is present in the list.
It checks for element presence by going through the list and comparing each value.
  - Checking Membership in a Dictionary
When you use the in keyword with a dictionary, it checks for key membership by default, not values.
It checks if the key exists in the dictionary, not whether the value exists.
